```python
# import needed libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# generate list of years for use in loop.
# year_list = [year for year in range(2010, 2020)]
# year_list
```

```python
# website URL to be scraped
url = 'https://www.imdb.com/search/title/?year=2010-01-01,2010-12-31&sort=boxoffice_gross_us,desc&view=advanced'
response = requests.get(url)
```

```python
# verify good response from website
response.status_code
```

```python
# use Beautiful Soup to parse website response
soup = BeautifulSoup(response.text)
```

```python
# find the collection of movie data in the response
movie_info_raw = soup.findAll('div')[46]
```

```python
# find the movie's name
movie_info_raw.findAll('h3')[0].findAll('a')[0].contents[0]
```

```python
# find the movie's MPAA rating
movie_info_raw.findAll('p')[0].findAll('span')[0].contents[0]
```

```python
# find the movie's genre
str(movie_info_raw.findAll('p')[0].findAll('span')[4].contents[0]).strip()
```

```python
# find the movie's gross box office (in millions)
float(str(movie_info_raw.findAll('p')[3].findAll('span')[4].contents[0]).strip().replace('$','').replace('M', ''))
```

```python
# loop through movie data and construct dict to hold the desired data
# loop variable i steps through each movie in the webpage response. There are 50 movies per display page.
# variable x is used to locate the proper 'p' tags, which increment by 4 for each movie entry.
# movie information is collected in a dict called 'movie_info'
x = 0
movie_info = {}
for i in range(0,50):
    movie_name = movie_info_raw.findAll('h3')[i].findAll('a')[0].contents[0]
    movie_mpaa = movie_info_raw.findAll('p')[x].findAll('span')[0].contents[0]
    movie_genre = str(movie_info_raw.findAll('p')[x].findAll('span')[4].contents[0]).strip()
    movie_gross = float(str(movie_info_raw.findAll('p')[x+3].findAll('span')[4]
                            .contents[0]).strip().replace('$','').replace('M', ''))
    movie_info[i+1] = {'title':movie_name, 'mpaa_rating':movie_mpaa, 'genre':movie_genre,
                     'gross_box_ofc':movie_gross, 'release_year':2010}
    x += 4
movie_info
```
